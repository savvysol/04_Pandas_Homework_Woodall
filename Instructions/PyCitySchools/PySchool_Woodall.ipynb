```python
import pandas as pd
```

```python
school_data_file = 'Resources/schools_complete.csv'
student_data_file = 'Resources/students_complete.csv'

schools = pd.read_csv(school_data_file)
schools.head()
```

```python
students = pd.read_csv(student_data_file)
students.head()
```

```python
students.columns
```

```python
old_cols = ['Student ID', 'student_name', 'gender', 'grade', 'school_name',
       'reading_score', 'math_score']

new_cols = ['SID', 'Name', 'Gender', 'Grade', 'School',
       'Reading', 'Math']

zip_cols = zip(old_cols,new_cols)
new_columns = dict(zip_cols)

new_students = students.rename(columns=new_columns)
students2 = new_students.set_index('Name')
students2.head()
```

```python
students3 = students2.drop(columns=['SID'])
students3.head()
```

```python
students3.describe()
```

```python
students4 = students3.groupby(['School','Grade','Gender','Name'],
                              axis=0,
                              sort=True,
                              observed=False,
                             ).sum()
students4
```

```python
students5 = pd.DataFrame(students4)
students5.describe()
```
